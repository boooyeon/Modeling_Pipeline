# Keras로 Linear&Logistic Regression 조금 더 맛보기

```python
import tensorflow as tf
from tensorflow import keras

import numpy as np
import pandas as pd
```

# Linear Regression on 쬐금 더 어려운 데이터

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
x = boston.data
y = boston.target

x.shape, y.shape
```

```python
# 모델링 : Sequential API
# 1번 청소 : 이미 만들어진 모델이 있다면 그 모델을 없애줘
keras.backend.clear_session()

# 2번 모델 선언
model = keras.models.Sequential()

# 3번 모델 블록 조립
model.add( keras.layers.Input(shape=(13,)) )
model.add( keras.layers.Dense(1) )

## 오리지널 Sequential API
# model.add( keras.layers.Dense(1, input_shape=(1,)) )

# 4번 컴파일 
model.compile(loss='mse', optimizer='adam')
```

```python
model.fit(x, y, epochs=10, verbose=1)
```

```python
# 예측
y_pred = model.predict(x)
y_pred[:5].reshape(-1)
```

```python
y[:5].reshape(-1)
```

# Logistic Regression 쬐금 더 어려운 데이터

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
x = cancer.data
y = cancer.target

# 모양 확인해볼 것
x.shape, y.shape
```

```python
# 혹시 이미 그려둔 그래프가 있다면 날려줘!
keras.backend.clear_session()

# model에 순차적으로 레이어를 쌓아가겠다는 의도!
model = keras.models.Sequential()

# model에 인풋 값을 받는 레이어를 넣음
model.add( keras.layers.Input(shape=(30,)) )
# model에 Dense 레이어를 넣을거야 (최초의 레이어) : weight를 곱하고, bias를 더해주는 과정
model.add( keras.layers.Dense(1, activation='sigmoid') )


# 오리지널 Sequential API
# model.add( keras.layers.Dense(1, input_shape=(1,), activation='sigmoid') )

# 컴파일 해주렴
model.compile(loss = 'binary_crossentropy',
              optimizer = 'adam',
              metrics=['accuracy'])
```

```python
# 데이터를 넣어서 학습시키자!
model.fit(x, y, epochs=10, verbose=1)
```

```python
# 결과 출력해줘!
print(y[15:])
print(model.predict(x[15:]))
```
