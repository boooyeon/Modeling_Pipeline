```python
import numpy as np
import pandas as pd

import tensorflow as tf
from tensorflow import keras

from sklearn.datasets import load_iris
```

## 데이터 불러오기

- iris.feature_names를 이용하여 판다스 데이터 프레임 만들기!

```python
iris = load_iris()
```

```python
x = iris.data
y = iris.target
```

```python
df_x = pd.DataFrame(x, columns=iris.feature_names)

df_x.head(3)
```

## train set, test set 구분하기

```python
from sklearn.model_selection import train_test_split
```

```python
train_x, test_x, train_y, test_y =\
    train_test_split(df_x, y, test_size=0.2, random_state=2022)
```

```python
train_x.shape, train_y.shape, test_x.shape, test_y.shape
```

## **length끼리, width끼리**

```python
df_x.columns
```

```python
tr_x_l = train_x.loc[: , ['sepal length (cm)', 'petal length (cm)'] ]
tr_x_w = train_x.loc[: , ['sepal width (cm)', 'petal width (cm)'] ]
```

```python
tr_x_l.shape, tr_x_w.shape
```

```python
tr_x_w.head(2)
```

```python
te_x_l = test_x.loc[: , ['sepal length (cm)', 'petal length (cm)'] ]
te_x_w = test_x.loc[: , ['sepal width (cm)', 'petal width (cm)'] ]
```

## One-hot Encoding

```python
train_y.shape
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y_len = len(set(train_y))
```

```python
train_y = to_categorical(train_y, y_len)
test_y = to_categorical(test_y, y_len)
```

```python
train_y.shape
```

## Modeling : multi-input & Concatenate layer

```python
## 모델링 : Functional API

# 1번 세션 클리어
keras.backend.clear_session()

# 2번 레이어 사슬처럼 엮기 : input layer 2개!
il_l = keras.layers.Input(shape=(2,) )
hl_l = keras.layers.Dense(2, activation='relu')(il_l)

il_w = keras.layers.Input(shape=(2,))
hl_w = keras.layers.Dense(2, activation='relu')(il_w)

cl = keras.layers.Concatenate()([hl_l, hl_w])
ol = keras.layers.Dense(3, activation='softmax')(cl)

# 3번 모델의 시작과 끝 지정
model = keras.models.Model([il_l, il_w] , ol)

# 4번 컴파일
model.compile(loss=keras.losses.categorical_crossentropy, metrics=['accuracy'],
             optimizer='adam')

# 양념
model.summary()
```

## 모델 시각화

```python
from tensorflow.keras.utils import plot_model
```

```python
plot_model(model, show_shapes=True)
```

## 학습 시켜보기!

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
es = EarlyStopping(monitor='val_loss',
                   min_delta=0,
                   patience=30,
                   verbose=1,
                   restore_best_weights=True)
```

```python
model.fit([tr_x_l, tr_x_w], train_y, validation_split=0.15,
          epochs=1000, verbose=1, callbacks=[es])
```

# IRIS - Connection I practice

- 위의 코드를 **그대로** 따라서 코드 작성하기!

## 데이터 불러오기

- iris.feature_names를 이용하여 판다스 데이터 프레임 만들기!

```python
iris = load_iris()

x = iris.data
y = iris.target
```

```python
df_x = pd.DataFrame(x, columns=iris.feature_names)
df_x.head()
```

## train set, test set 구분하기

```python
from sklearn.model_selection import train_test_split
```

```python
train_x, test_x, train_y, test_y =\
    train_test_split(df_x, y, test_size=0.2, random_state=2022)
```

```python
train_x.shape, train_y.shape
```

## **length끼리, width끼리**

```python
train_x.columns
```

```python
tr_x_l = train_x[['sepal length (cm)', 'petal length (cm)']]
tr_x_w = train_x[['sepal width (cm)', 'petal width (cm)']]
```

```python
te_x_l = test_x[['sepal length (cm)', 'petal length (cm)']]
te_x_w = test_x[['sepal width (cm)', 'petal width (cm)']]
```

```python
tr_x_l.shape, tr_x_w.shape
```

## One-hot Encoding

```python
from tensorflow.keras.utils import to_categorical
```

```python
y_len = len(set(train_y))
```

```python
train_y = to_categorical(train_y, y_len)
test_y = to_categorical(test_y, y_len)
```

```python
train_y.shape
```

## Modeling : multi-input & Concatenate layer

```python
from tensorflow.keras.backend import clear_session
from tensorflow.keras.models import Model

from tensorflow.keras.layers import Input, Dense, Concatenate
from tensorflow.keras.losses import categorical_crossentropy
from tensorflow.keras.optimizers import Adam

from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.utils import plot_model
```

```python
## 모델링 : Functional API

# 1. 세션 클리어
clear_session()

# 2. 레이어 사슬처럼 엮기
il_w = Input(shape=(2,), name='input_width')
hl_w = Dense(2, activation='relu', name='hidden_width')(il_w)

il_l = Input(shape=(2,), name='input_length')
hl_l = Dense(2, activation='relu', name='hidden_length')(il_l)

cl = Concatenate(name='concat_layer')([hl_w, hl_l])
ol = Dense(3, activation='softmax')(cl)

# 3. 모델의 시작과 끝 지정
model = Model([il_w, il_l], ol)

# 4. 컴파일
model.compile(loss=categorical_crossentropy, metrics=['accuracy'],
             optimizer=Adam(0.01) )

model.summary()
```

## 모델 시각화

```python
plot_model(model, show_shapes=True, show_layer_names=True)
```

## 학습 시켜보기!

```python
es = EarlyStopping(monitor='val_loss', min_delta=0, verbose=1, patience=5, restore_best_weights=True)
```

```python
model.fit([tr_x_w,tr_x_l], train_y, verbose=1, callbacks=[es], validation_split=0.2, epochs=1000)
```
